### Exercise 1

Create box plots on a single set of axes for all five experiments in the morley data set.

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/morley.csv')
# Dropping the "Unnamed column"
df = df.drop(columns=['Unnamed: 0'])
# Checking if ther is any null value
df.isnull().values.sum()
```

```python
df
```

```python
# Separating the Experiemnts for the box plots
exp_1 = df.loc[df['Expt'] == 1]
exp_2 = df.loc[df['Expt'] == 2]
exp_3 = df.loc[df['Expt'] == 3]
exp_4 = df.loc[df['Expt'] == 4]
exp_5 = df.loc[df['Expt'] == 5]
```

```python

fig, axs = plt.subplots(1,5,figsize=(15, 10))

# Experiment 1 Plot
axs[0].boxplot(exp_1['Speed'])
axs[0].set_title('Experiment 1', size=15)

# Ploting red dots for each value inside exp_1
axs[0].plot(np.ones(exp_1['Speed'].shape), exp_1['Speed'], 'r.')

# Plotting the mean value 
axs[0].plot([1.0], [np.mean(np.array(exp_1['Speed']))], 'go')



# Experiment 2 Plot
axs[1].boxplot(exp_2['Speed'])
axs[1].set_title('Experiment 2', size=15)

# Ploting red dots for each value inside exp_2
axs[1].plot(np.ones(exp_1['Speed'].shape), exp_2['Speed'], 'r.')

# Plotting the mean value 
axs[1].plot([1.0], [np.mean(np.array(exp_2['Speed']))], 'go')


# Experiment 3 Plot
axs[2].boxplot(exp_3['Speed'])
axs[2].set_title('Experiment 3', size=15)

# Ploting red dots for each value inside exp_3
axs[2].plot(np.ones(exp_3['Speed'].shape), exp_1['Speed'], 'r.')

# Plotting the mean value 
axs[2].plot([1.0], [np.mean(np.array(exp_3['Speed']))], 'go')


# Experiment 4 Plot
axs[3].boxplot(exp_4['Speed'])
axs[3].set_title('Experiment 4', size=15)

# Ploting red dots for each value inside exp_4
axs[3].plot(np.ones(exp_3['Speed'].shape), exp_4['Speed'], 'r.')

# Plotting the mean value 
axs[3].plot([1.0], [np.mean(np.array(exp_4['Speed']))], 'go')



# Experiment 5 Plot
axs[4].boxplot(exp_5['Speed'])
axs[4].set_title('Experiment 3', size=15)

# Ploting red dots for each value inside exp_5
axs[4].plot(np.ones(exp_3['Speed'].shape), exp_5['Speed'], 'r.')

# Plotting the mean value 
axs[4].plot([1.0], [np.mean(np.array(exp_5['Speed']))], 'go')
```

```python
# Reading the dataset
iris_df = pd.read_csv("https://gist.github.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
# Checking if ther is any null value
iris_df.isnull().values.sum()
```

```python
# looking at the df to see the name of the columns 
iris_df
```

```python
# Breaking down the numerical variables per flower type
setosa = iris_df.loc[iris_df['variety'] == "Setosa"]
virginica = iris_df.loc[iris_df['variety'] == "Virginica"]
versicolor = iris_df.loc[iris_df['variety'] == "Versicolor"]
```
