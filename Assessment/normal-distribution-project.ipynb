# Normal Distribution

Normal Distribution, also known as Gaussian or Gauss distribution is a continuous probability distribution where the data near to the mean occur more frequently than the data far from the mean. When plotted, normal distribution will display a curve called “bell curve” due to its shape. The normal distribution is a type of symmetrical distribution. Symmetrical distribution are distribution where the <i>mean</i>, <i>median</i>, and <i>mode</i>  have the same value and occur at the same spot and represent the highest point of the distribution.

## Parameters of the normal distribution:

**Mean:** The mean represents the highest point in the normal distribution graph, its peak and most of the values or data points are clustered around it. Changing the value of the mean will cause the graph representation of the normal distribution to move its curve either to the right if the value increases or to the left of its value decreases.


```python
## Example of mean effect on a normal distribuition

#Mean and standart deviation
mu, sigma = 0.0 , 1.0

## Creating values for X
x = np.linspace(-2, 7, 1000)

# Probability Density Function:


for mu in [1.0, 2.0, 3.0, 4.0]:
    plt.plot(x, (1.0 / np.sqrt(2 * np.pi * sigma**2) * np.exp(-((x - mu)**2 / (2 * sigma**2)))), label= rf'$\mu = {mu}$')
    # Mean line for each function
    plt.axvline(mu, color="red",linestyle="--")
plt.legend()
```

**Standard deviation:** represent the dispersion and how far a data point is from the mean, when visualised the standard deviation dictates the shape of the curve. Larger standards deviations causes the bell curve to look flat, on the other hand a small value for the standard deviation will cause the curve to look steep and make the data more clustered around the mean.

We can see below a vissual representation of this effect in place.

```python
## Example of mean effect on a normal distribuition shape

#Mean and standart deviation
mu, sigma = 0.0 , 1.0

## Creating values for X
x = np.linspace(-4, 4, 1000)

for sigma in [0.5, 0.8, 1.0, 1.5]:
    plt.plot(x, (1.0 / np.sqrt(2 * np.pi * sigma**2) * np.exp(-((x - mu)**2 / (2 * sigma**2)))), label= rf'$\sigma = {sigma}$')
    plt.axvline(mu, color="red",linestyle="--")
plt.legend()
```

## Probability density function

The normal distribution probability function is the following:  $F(x) = \frac{1}{\sqrt{2 \pi \sigma^2 }} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $. 

This is the function used to plot the normal distribuition. It was written down as a python code on the examples above as <code>(1.0 / np.sqrt(2 * np.pi * sigma**2) * np.exp(-((x - mu)**2 / (2 * sigma**2))))</code>. Where mu is the mean and sigma the standard deviation. For us to calculate $F(x)$ we need to know the mean and the standard deviation. below we are going to calculate it for x = 0, mu = 0 and sigma = 1

$F(x) = \frac{1}{\sqrt{2 \pi \sigma^2 }} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $. &rarr; $F(0) = \frac{1}{\sqrt{2 \pi*1^2 }} e^{-\frac{(0 - 0)^2}{2*1^2}} $. &rarr; $F(0) = \frac{1}{\sqrt{2 \pi*1^2 }} e^0 $. &rarr; $F(0) = \frac{1}{\sqrt{2 \pi}} $. &rarr; $F(0) = \frac{1}{2.5066} $. &rarr; $F(0) =0.3989$.

#### Scipy Probability Density Function

For comparison let’s use the SciPy module to calculate it and see if the result matches. The PDF takes the same three parameters $x=0$ ,  $\mu=0$ ,  $\sigma=1$

```python
# Calculating the F(x) for x=0 using the Scipy normal distribuition PDF
# The function takes, x as first parameter, mu as the second one and the sigma as third.
x = 0
mu = 0
sigma = 1
fx = stats.norm.pdf(x, mu, sigma)
print(f"F(x) is equal to: {fx}")
```

As we can see above, in python there are other ways to calculate and plot a normal distribution. Some packages such as SciPy and NumPy can be used for that. Now let’s plot a normal distribution probability density function using Python's SciPy package. We are going to use the norm.rvs method that will generate a normal distribuition for random values of x.

```python
# Setting the values for the mean and standard deviation
mu = 0
sigma = 1
x = np.linspace(-4, 4, 1000)
# Calling Scipy norm.rvs to create a normal distribuition with random values for X
norm_random = stats.norm.pdf(x,loc=mu, scale=sigma)
plt.plot(x, norm_random)
plt.show()
```

## Proprieties of the normal distribution:

**Symmetry**:
If you divide it in the middle you will be left with two equals halves with one-half of the datapoints on each side.

**Mean, Median and Mode:**

- **Mean:** The mean represents the highest point in the normal distribution graph, its peak and most of the values or data points are clustered around it.

- **Mode:** is the most common score, in the distribution, and since the most common score is the one that occurs at the peak of the distribution and the mean also represents the highest point in the distribution, therefore the mean and the mode are the same value in the normal distribution.

- **Median:** is the score positioned in the middle of the dataset when it is ordered in an ascending way and since the normal distribution is symmetrical, meaning that the middle point is the highest value of the normal distribution curve and also the same value as the mean and the mode.

## Empirical Rule

The empirical rule or 68-95-99.7% rule, states that between -1 σ and +1 σ holds 68% of the distribution data. If we increase the standard to -2σ and +2σ the data covered for this interval will be 95% of the whole data. And if we go even further and consider -3 and +3 standards deviations, we will cover 99.7%, for 4σ or higher there is only a probability of 0.3% of a value to be in this range. Because of that, in general,  the 3σ point is the one considered to hold the most significant probability values.

## Standard normal distribution and Z scores

-	**Standard normal distribution** is a normal distribution where the mean is 0 and the standard deviation is 1. It is used together with z scores to determine the probability of data points to be found in the distribution.
-	**Z – Score*** is used to convert and compare results from normally distributed samples to a standardize distribution. It shows how many standards deviations a datapoint is from the mean. 


### Z Score Formula

-	To calculate a z score from a sample we need to know the mean and standard deviation. The formula to calculate the Z score we subtract the mean of the sample from the value of the X (which is our datapoint value) and divide it by the standard deviation of the sample. The formula to calculate the Z score is:   

$$Z = \frac{X - \mu}{\sigma} $$

For example, suppose a male student whose height is 190 cm wants to know how tall or shorter he is in relation to an average male. After some research he found out that the average men height is [178 cm](https://ourworldindata.org/human-height) and it has a standard deviation of 8 cm. We can find it out using the Z score.

$Z = \frac{X - \mu}{\sigma} $ &rarr; $Z = \frac{190 - 178}{8} $ &rarr; $Z = \frac{12}{8} $ &rarr; $Z = 1.5 $

Looking at the Z score above, it shows that the student is 1.5 standards deviation above the mean. This result shows him that he is above the average but how much above the average is he, how many percent of the population is shorter than him? As a example, if his Z score was 2, we could safely assume that he would be taller than 97.5% of our sample population just by looking at the empirical rule graph. But since his Z score is 1.5, there is other ways to find it out. 

#### Z Score Table

We can use the [Z score table](https://math.fandom.com/wiki/Z_table_(cumulative)) to find it.

![Z Table](https://www.z-table.com/uploads/2/1/7/9/21795380/8573955.png)

The table above is the cumulative probability for Z score, it shows the probability t the left of the Z score. That means that the student with a Z score of 1.5 is taller than 0.9332 of the data population. Since the whole area of the normal distribution is 1  we can multiply it for 100 and get that the student is taller than 93.32% of the data population.

#### Scipy Cumulative Density Function

Scipy normal module also have a method to deal with Z scores and cumulative functions. It’s the norm.cdf, for this method you can give it the parameters as same as the probability density function, however it will plot a cumulative density function. We can see below a example of the cumulative density function.

```python
# Checking our Z score calculation against the Scipy norm.cdf
stats.norm.cdf(1.5)
```

```python
# Ploting the cumulative densit function for the student case

# Setting the values for the mean and standard deviation and student height
student = 190
mu = 178
sigma = 8

# Creating value for X
x = np.linspace(150, 210, 1000)

# Calling Scipy norm.cdf to create a cumulative density function
cdf_student = stats.norm.cdf(student,loc=mu, scale=sigma)
cdf = stats.norm.cdf(x,loc=mu, scale=sigma)
plt.plot(x, cdf)
# lines for the value of Student height and probability.
plt.axvline(student, color="green",linestyle="-")
plt.axhline(cdf_student, color="green",linestyle="-")

plt.ylabel('Cumulative Probability')
plt.xlabel('Height')
plt.show()
```

## Normal Distribuition and Central Limit Theorem 

Central limit theorem is a statistical concept that implies that methods applied to normal distribution can also be applied to problems related to other types of distributions.
That happens, because according to the theorem the means of random samples from a distribution can be assumed to be normally distributed, given enough number of samples.


This can be seen below, where a exponential distribution was plotted as an example 

```python
#Exponential distribution Plot
exp = stats.expon.rvs(loc=20, scale=3, size=10000)
sns.histplot(exp)
```

Now if we get samples from the exponential distribution and calculate their means, we can plot those means and it will look a lot like a normal distribuition.

```python
# Generating 1000 random exp distribution
count = 5000
sample_mean = 0
means = []
for i in range(count):
    
    exp = stats.expon.rvs(loc=20, scale=3, size=300)
    # Getting the mean from the sample
    mean = exp[30:220].mean()
    means.append(mean)

# Plotting the means
sns.histplot(means, kde=True)
```
